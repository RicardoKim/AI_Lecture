# 1. 클래스란?

클래스는 여러 함수들을 모아놓은 틀 같은겁니다. 다음 예시를 보시죠


````python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
````

1. 다음과 같이 여러 함수들을 FourCal이라는 클래스로 묶었습니다. 여기서 우리는 FourCal을 클래스라고 합니다. 

2. 클래스안에 정의되어있는 함수는 메소드라고 불립니다.




```
def setdata(self, first, second):
        self.first = first
        self.second = second
```



setdata는 여러가지 변수를 받는 함수입니다. 여기서 우리는 변수를 두가지로 나눌 수 있습니다.

- 객체변수 : first, second를 객체변수라고 부르며 일반적으로 함수에 넣어져서 연산에 사용되는 변수를 의미합니다.

- 매개변수 : self를 매개변수라고 부르며 전달인자의 역할을 합니다. 무슨 뜻이냐 하면 메소드들이 특정 변수를 공유하기 위해서 self를 이용해서 변수를 공유한다는 뜻입니다.

 위의 예제에서는 setdata에서 받아온 first라는 변수를 다른 메소드들이 사용하게 하기 위해서 self.first라는 매개변수를 사용하는 변수로 저장해놓고 다른 변수들은 self인 매개변수만 받아와서 이후 self.first를 호출해서 사용하게 됩니다.

````python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
````

````python
class InitFourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
````

# 4. 메소드 오버라이딩

오버라이딩이라는 것은 우리가 부모 클래스의 메소드 중에 특정 메소드를 자식 클래스에서 새롭게 정의하는 것이다.

다음 예시를 봐보자

````python
class SafeFourCal(InitFourCal):
    pass
````

````python
Wrong_example = SafeFourCal(4, 0)
Wrong_example.div()
````

부모 클래스에 정의된 나눗셈은 분모가 0일때는 고려하지 않고 만든 것이다. 따라서 우리는 위의 예제를 돌아가게 하려면 새롭게 div 메소드를 정의해야한다.

````python
 class SafeFourCal(InitFourCal):
    def div(self):
        if self.second == 0:  
            return 0
        else:
            return self.first / self.second
````

````python
Fixed_example = SafeFourCal(4, 0)
Fixed_example.div()
````

다음 ElementwiseCaculation을 상속 받아서 mul이라는 메소드를 원소 곱에서 행렬곱을 하는 메소드로 바꿔보세욥

````python
import numpy as np
````

````python
class ElementwiseCalculation():
    def __init__(self):
        self.x = np.random.normal(size = (3,3))
        self.y = np.random.normal(size = (3,1))
    
    def mul(self):
        output = np.multiply(self.x, self.y)
        return output
````

````python
class MatrixCalculation(ElementwiseCalculation):
    
    def mul(self):
        output = np.matmul(self.x, self.y)
        return output
````

````python
Elementwise = ElementwiseCalculation()
Matrixcal = MatrixCalculation()
````

````python
print(Elementwise.mul().shape)
print(Matrixcal.mul().shape)
````

## 5.1 Dataset 클래스 살펴보기

````python
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import pandas as pd
````

````python
df_train = pd.read_csv('./sample_data/mnist_train_small.csv', header=None)
df_test = pd.read_csv('./sample_data/mnist_test.csv', header=None)

train_labels = df_train.iloc[:, 0]
train_images = df_train.iloc[:, 1:]
test_labels = df_test.iloc[:, 0]
test_images = df_test.iloc[:, 1:]
````

다음은 우리가 데이터셋을 정의할때 상속 받는 torch에 dataset이라는 클래스이다.

image.png

````python
class MNISTDataset(Dataset):
    def __init__(self, images, labels=None):
        self.X = images
        self.y = labels

         
    def __len__(self):
        return (len(self.X))
    
    def __getitem__(self, i):

        data = self.X.iloc[i, :]
        data = np.asarray(data).astype(np.float32).reshape(28, 28, 1)
        
        if self.y is not None:
            return (data, self.y[i])
        else:
            return data
````

- 1. *__len*__ 의 경우 파이썬에 내장되어있는 len()함수의 뿌리 같은 메소드이다. 우리가 len()을 했을 때는 물론 변수의 길이를 반환하지만 이 또한 *__len*__로 정의 되어있는 함수를 호출하는 api에 불과하다.

- 2. *__getitem*__의 경우 우리가 객체에 변수를 직접 접근하지 않고 리스트를 슬라이싱 하기 위해서 선언되는 함수이다. 즉, 우리가 특정 클래스가 가지고 있는 변수를 슬라이싱 해서 특정 값을 반환 받고 싶을 때 직접적으로 클래스 변수에 접근해서 변수를 슬라이싱 하지 않고 해당 메소드를 호출해서 값을 반환받는 것이다.

````python
train_data = MNISTDataset(train_images, train_labels)
test_data = MNISTDataset(test_images, test_labels)
````

````python
train_loader = DataLoader(train_data, batch_size=64, shuffle=True, drop_last = True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=True, drop_last = True)
````

## 5.2 nn.Module 살펴보기 

````python
import torch.nn as nn
````

````python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 10),
            nn.ReLU()
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
````

## super란?

````python
class father():  # 부모 클래스
    def handsome(self):
        print("잘생겼다")
 
 
class brother(father):  # 자식클래스(부모클래스) 아빠매소드를 상속받겠다
    '''아들'''
 
 
class sister(father):  # 자식클래스(부모클래스) 아빠매소드를 상속받겠다
    def pretty(self):
        print("예쁘다")
 
    def handsome(self):
        '''물려받았어요'''
 
 
brother = brother()
girl = sister()
````

````python
brother.handsome()
````

````python
girl.handsome()
````

````python
girl.pretty()
````

분명 sister의 경우도 father를 상속받았는데 왜 handsome 메소드가 작동하지 않을까요?

사실 이 부분은 당연한 부분인데 handsome이 father클래스의 handsome 메소드에서 오버라이드 되서 아무것도 출력이 안되었기 때문이다. sister를 다음과 같이 수정하면 작동이 되긴 된다.



```
class sister(father):  
    def pretty(self):
        print("예쁘다")

```



하지만 우리는 handsome이라는 method를 오버라이드 하여 내용을 더 추가하고 싶다면? 저런식으로 하면 father 클래스를 상속 받을 이유가 없게 됩니다. 그냥 처음부터 클래스를 짜는게 맞겠죠.

이럴때 super를 사용하게 됩니다. 아래 예시를 보시죠

````python
class sister(father):  
    def pretty(self):
        print("예쁘다")
 
    def handsome(self):
        super().handsome()
        print("아빠 덕분")
````

````python
new_girl = sister()
````

````python
new_girl.handsome()
````

이제 우리는 위에서 언급된 부분을 이해할 수 있는 준비가 되었다.

````python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 10),
            nn.ReLU()
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
````

우리가 추가하고자 하는거는 다음 NeuralNetwork의 클래스에서 사용할 레이어들을 *__init*__에서 정의하고자 하는 것이다.
